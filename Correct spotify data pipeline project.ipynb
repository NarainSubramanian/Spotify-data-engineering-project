```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
# here the spotify client id and the client secret id can be obtained from "spotify for developers"
client_credentials = SpotifyClientCredentials(client_id = '<client_id>',client_secret = '<client_secret>')
```

```python
sp = spotipy.Spotify(client_credentials_manager = client_credentials)
```

```python
data = sp.search("Anirudh")   # can also type playlist_ID here
```

```python
album_list = []
items = data['tracks']['items']
for track in items:
    album_id = track['album']['id']
    album_name = track['album']['name']
    album_release_date = track['album']['release_date']
    album_total_tracks = track['album']['total_tracks']
    album_url = track['album']['external_urls']['spotify']
    


    
    
    # Store in dictionary
    album_data = {
        'album_id': album_id,
        'album_name': album_name,
        'album_release_date': album_release_date,
        'album_total_tracks': album_total_tracks,
        'album_url': album_url,
        
    }
    album_list.append(album_data)

# Print the result
for album in album_list:
    print(album)
```

```python
artist_list = []
for track in items:
    for artist in track['artists']:
        artist_name = artist['name']
    
    for artist in track['artists']:
            artist_id = artist['id']

    track_name = track['name']  # Track name

    for artist in track['artists']:
        external_url = artist['href']

    # Store in dictionary
    artist_data = {
        'artist_name': artist_name,
        'track_name': track_name,
        'artist_id': artist_id,
        'external_url':external_url
    }
    artist_list.append(artist_data)

# Print the result
for artist in artist_list:
    print(artist)
```

```python
song_list = []
for row in items:
    song_id = row['id']
    song_name = row['name']
    song_duration = row['duration_ms']
    song_url = row['external_urls']['spotify']
    song_popularity = row['popularity']
    

    song_element = {
        'song_id' : song_id,
        'song_name' : song_name,
        'duration_ms' : song_duration ,
        'url' : song_url,
        'popularity' : song_popularity
    }

    song_list.append(song_element)

# Print the result
for song in song_list:
    print(song)
```

```python
import pandas as pd

album_df = pd.DataFrame.from_dict(album_list)     #creates df for albums
album_df = album_df.drop_duplicates(subset = ['album_id'])  # removes duplicates from album df

artist_df = pd.DataFrame.from_dict(artist_list)   # creates df for artists
artist_df = artist_df.drop_duplicates(subset = ['artist_id'])  # removes duplicates from artist df

song_df = pd.DataFrame.from_dict(song_list)       #song dataframe
```
